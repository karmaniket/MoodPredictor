```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
import joblib as jl
```

```python
df = pd.read_csv('mood_dataset.csv')
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['mood'], test_size=1, random_state=42)
```

```python
vctr = TfidfVectorizer()
```

```python
X_train_vec = vctr.fit_transform(X_train)
```

```python
X_test_vec = vctr.transform(X_test)
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train_vec, y_train)
```

```python
jl.dump(model, 'model.pkl')
```

```python
jl.dump(vctr, 'vectorizer.pkl')
```

```python
y_pred = model.predict(X_test_vec)
```

```python
print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
```

```python
print(classification_report(y_test, y_pred, zero_division=0))
```
